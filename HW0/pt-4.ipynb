```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread('chest-xray.png', cv.IMREAD_UNCHANGED)
```

```python
img.shape
```

### It is already grayscale

```python
type(img)
```

```python
plt.imshow(img, cmap = 'gray', vmin = 0, vmax = 255)
```

```python
img.dtype
```

```python
print(f'Total memory used by the grayscale image: {img.nbytes / 1000} KB')
```

```python
cropped_gray_image = img[:, :img.shape[1] // 2]
```

```python
plt.imshow(cropped_gray_image, cmap = 'gray', vmin = 0, vmax = 255)
```

```python
flipped_x_axis = cv.flip(cropped_gray_image, 0)
```

```python
plt.imshow(flipped_x_axis, cmap = 'gray', vmin = 0, vmax = 255)
```

```python
fig, ax = plt.subplots(1, 3, figsize = (10, 5))

ax[0].imshow(img, cmap='gray', vmin = 0, vmax = 255)
ax[0].set_title('Gray image')

ax[1].imshow(cropped_gray_image, cmap='gray', vmin = 0, vmax = 255)
ax[1].set_title('Left chest')

ax[2].imshow(flipped_x_axis, cmap='gray', vmin = 0, vmax = 255)
ax[2].set_title('Flipped left chest')

plt.show()
```
